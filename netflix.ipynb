```python
netflix_movies=netflix_df[['title','country','genre','release_year','duration']]
netflix_movies
```

```python
short_movies=netflix_movies[netflix_movies['duration'] < 60]
short_movies
```

```python
colors=[]
for index,row in netflix_movies.iterrows():
    if row['genre']=="Çhildren":
        colors.append('red')
    elif row["genre"]=='Documentaries':
        colors.append('blue')
    elif row["genre"]=='Stand-Up':
        colors.append('grey')
    else:
        colors.append('green')
fig,ax=plt.subplots()
scatter=ax.scatter(netflix_movies['release_year'],netflix_movies['duration'],c=colors)
ax.set_xlabel('Release year')
ax.set_ylabel('Duration (min)')
ax.set_title("Movie Duration By Year Of Release")
legend_labels=['Children',"Documentaries","Stand-Up","Others"]
ax.legend(handles=scatter.legend_elements()[0],labels=legend_labels)

plt.show()
```
